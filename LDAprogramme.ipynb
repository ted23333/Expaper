```python

anatext = pd.read_csv("resultcyt.csv",header=None)
stoplist = 'stoplist.txt'
stop = pd.read_csv(stoplist,header=None,sep ='tipdm',engine='python')
stop = [" ",""] + list(stop[0])
# 定义分隔函数，然后用apply进行广播
anatext[1] = anatext[0].astype(str).apply(lambda s: s.split(" "))
anatext[2] = anatext[1].apply(lambda x: [i for i in x if i not in stop])
#  怎么建立语料库？？？
 # 建立词典
stop_dict = corpora.Dictionary(anatext[2])
# 建立语料库
stop_corpus = [stop_dict.doc2bow(i) for i in anatext[2]]
```

```python
# LDA模型训练
LDA_model = models.LdaModel(stop_corpus,num_topics = 10,id2word = stop_dict)
for i in range(10):
    print(LDA_model.print_topic(i))
```
