```python
for i in range(0,403189):
    try:
        text['标题／微博内容'][i] = re.sub(r'【.*】',"",text['标题／微博内容'][i])
    except TypeError:
        print(i)
        print('发生了错误')
    # 去掉【】内容
```

```python
for i in range(0,403189):
    try:
         text['标题／微博内容'][i] = re.sub(r'/',"",text['标题／微博内容'][i])
             # 去掉/符号
         text['标题／微博内容'][i] = re.sub(r'@.*:',"",text['标题／微博内容'][i])
        # 去掉@...:
         text['标题／微博内容'][i] = re.sub(r'@.*\s$',"",text['标题／微博内容'][i])
    # 去掉@+空格
         text['标题／微博内容'][i] = re.sub(r'[?？！!]',"",text['标题／微博内容'][i])
    # 去掉特殊符号
         text['标题／微博内容'][i] = re.sub(r'http.*\s$',"",text['标题／微博内容'][i])
    # 去掉http链接
         text['标题／微博内容'][i] = re.sub(r'@.*$',"",text['标题／微博内容'][i])
    # 去掉@的结尾
    except TypeError:
        print(i)
        print('发生了错误')
```

```python
text.to_csv('result.csv')
```
